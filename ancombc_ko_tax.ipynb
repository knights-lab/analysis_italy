```r
phylo_object <- subset_samples(phylo_object, is.finite(Age))
```

```r
phylum_data = aggregate_taxa(phylo_object, "Phylum")
```

```r
out = ancombc(phyloseq = phylum_data, formula = "Age + OUTCOME + Hospital_bin", 
              p_adj_method = "holm", zero_cut = 0.90, lib_cut = 1000, 
              group = "Hospital_bin", struc_zero = TRUE, neg_lb = TRUE, tol = 1e-5, 
              max_iter = 100, conserve = TRUE, alpha = 0.05, global = TRUE)

res = out$res
res_global = out$res_global
```

```r
tab_coef = res$beta
```

```r
tab_p = res$p_val

tab_p
```

```r
samp_frac = out$samp_frac
# Replace NA with 0
samp_frac[is.na(samp_frac)] = 0 

# Add pesudo-count (1) to avoid taking the log of 0
log_obs_abn = log(abundances(phylum_data) + 1) 
# Adjust the log observed abundances
log_obs_abn_adj = t(t(log_obs_abn) - samp_frac)
# Show the first 6 samples
round(log_obs_abn_adj[, 1:6], 2)
```
